## Analyse

### Durch deine Paarplot-Matrix werden Zusammenhänge zwischen einzelnen Parametern leicht sichtbar

```python
sns.pairplot(data, palette=color_dict, vars = ['Anteil Sozialhilfeempfänger', 'Gymnasialquote', 'Altersquotient', 'Anteil Grünflächen', 'Wohnfläche pro Person', 'Vermögenssteuer pro Veranlagung', 'Erwerbstätigenquote', 'Anteil Personen ohne Religionszugehörigkeit'], hue = 'Wohnviertel Name')
```

Da *Anteil Personen ohne Religionszugehörigkeit* und *Arbeitslosenquote* nur bis 2019 erfasst sind, arbeite ich mit den Daten von davor.

```python
data['size'] = data.apply(lambda row : (row['Publikationsjahr'] - 2000), axis=1)
data['Religionszugehoerig'] = data.apply(lambda row : (1 - row['Anteil Personen ohne Religionszugehörigkeit']), axis=1)

```

```python
data_2022 = data.loc[data['Publikationsjahr'] == 2022]
data_2021 = data.loc[data['Publikationsjahr'] == 2021]
data_2019 = data.loc[data['Publikationsjahr'] == 2019]

#print(data_2022.head())
data_2019.describe()
```

### Quartiere nach Altersquotient

```python
plt.figure(figsize = (15, 12))
plt.title('Wo wohnen die Älteren? Daten von 2019')
plt.xticks(rotation=90)
sns.barplot(data = data_2019, x = 'Wohnviertel Name', y = 'Altersquotient', palette=color_dict, order=data_2019.sort_values('Altersquotient',ascending=False)['Wohnviertel Name'])
sns.despine()
```

### Quartiere nach % Religionszugehörigkeit

```python
plt.figure(figsize = (15, 12))
plt.title('Religionszugehörigkeit nach Quartier, Daten von 2019')
plt.xticks(rotation=90)
sns.barplot(data = data_2019, x = 'Wohnviertel Name', y = 'Religionszugehoerig', palette=color_dict, order=data_2019.sort_values('Religionszugehoerig',ascending=False)['Wohnviertel Name'])
sns.despine()
```

## Gibt es einen Zusammenhang zwischen Alter der Bevölkerung und Höhe der Religionszugehörigkeit?

```python
plt.figure(figsize = (15, 12))
plt.title('Religionszugehörigkeit nach Alter der Quartiersbewohnenden, 2019')
p1 = sns.scatterplot(data = data_2019, x = 'Jugendquotient', y = 'Religionszugehoerig', hue = 'Wohnviertel Name', s=100, legend=False, palette=color_dict)

#sns.regplot(data = data, x = 'Jugendquotient', y = 'Religionszugehoerig')
def label_point(x, y, val, ax):
    a = pd.concat({'x': x, 'y': y, 'val': val}, axis=1)
    for i, point in a.iterrows():
        ax.text(point['x'], point['y']+0.003, str(point['val']), horizontalalignment='center')

label_point(data_2019['Jugendquotient'], data_2019['Religionszugehoerig'], data_2019['Wohnviertel Name'], plt.gca())  


sns.despine()
```

Menschen in Quartieren mit höhrem Jugendanteil sind eher religiös. Über die letzten Jahren hat aber in den meisten Wohnvierteln die Religionszugehörigkeit abgenommen:

```python
plt.figure(figsize = (15, 12))
plt.title('Religionszugehörigkeit nach Alter, Veränderung über die Jahre')
sns.scatterplot(data = data, x = 'Jugendquotient', y = 'Religionszugehoerig', hue = 'Wohnviertel Name', size = 'size', palette=color_dict)
sns.despine()
```

## Welches Quartier hatte die grösste Veränderung in Jugendquotient und Altersquotient?


Neuen DataFrame ausschliesslich mit den relevanten Spalten bauen

```python
data_sub1 = data.iloc[:,[0,1,2,8,10]]

data_2015 = data_sub1[(data_sub1['Publikationsjahr'] == 2015)]
data_2022 = data_sub1[(data_sub1['Publikationsjahr'] == 2022)]

data_2022 = data_2022.rename(columns={'Altersquotient': 'AQ_2022', 'Jugendquotient': 'JQ_2022'})
data_2015 = data_2015.rename(columns={'Altersquotient': 'AQ_2015', 'Jugendquotient': 'JQ_2015'})

data_2015 = data_2015.iloc[:,[1,2,3,4]]
data_2022 = data_2022.iloc[:,[1,2,3,4]]

data_sub1 = pd.merge(data_2015, data_2022, how='left', on=['Wohnviertel ID', 'Wohnviertel Name'])
```

```python
data_sub1['JQ_Change'] = (data_sub1['JQ_2022']-data_sub1['JQ_2015'])/data_sub1['JQ_2015'] * 100
data_sub1['AQ_Change'] = (data_sub1['AQ_2022']-data_sub1['AQ_2015'])/data_sub1['AQ_2015'] * 100

data_sub1 = data_sub1.sort_values('JQ_Change', ascending=False)
data_sub1.head(3)
```

Bettingen hatte von 2015 zu 2022 einen Zuwachs an Jugendlichen von einem Drittel, gefolgt von der Altstadt Grossbasel und den Vorstädten.

```python
data_sub1 = data_sub1.sort_values('AQ_Change', ascending=False)
data_sub1.head(3)
```

Den grössten Zuwachs an Älteren hatten von 2015 zu 2022 Matthäus und Kleinhüningen. Während im Matthäusquartier gleichzeitig der Jugendquotient abnahm, gab es in Kleinhüningen einen Zuwachs in beide Richtungen.

Beide Quartiere haben einen vergleichsweise niederen Altersquotienten.
